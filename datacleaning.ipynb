### HANDLING MISSING VALUES

```python
missing_vals = df[['Age', 'Cabin', 'Fare']].isnull().sum()
```

```python
for i in missing_vals:
    print(f"The missing values are {i} and missing values percent is {(i/len(df)) * 100} ")
```

#### 1) THE FARE COLUMN HAS 0.2% MISSING VALUES THUS WE CAN REMOVE THE MISSING ROW

```python
# Removed 1 missing val and saved in original dataset

df = df.dropna(subset=['Fare'])
```

```python
df['Fare'].isna().sum()
```

#### 2) THE AGE COLUMN HAS 20% MISSING VALUES THUS WE CANNOT ENTIRELY REMOVE ROWS BUT FILL THE VALUES WITH MEAN OF AGE

```python
# Filling the Age col with mean

mean_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_age)
```

```python
df['Age'].isnull().sum()
```

```python
df.shape
```

#### 3) THE CABIN COLUMN HAS 78% MISSING VALS THUS CREATING FEATURE (HAS_CABIN) AND DROPING COLUMN

```python
df['Has_Cabin'] = df['Cabin'].notnull().astype(int)
```

```python
df = df.drop('Cabin', axis=1)
```

```python
df.head()
```

```python
df['Has_Cabin'].value_counts()
```

### HANDLING INCONSISTENCIES
#### THE NAME AND TICKET COLUMN HAS INCONSISTENCY

```python
# starting with Name col

df['Name']
```

```python
df[['Surname', 'Name']] = df['Name'].str.split(',', expand=True)
```

```python
df.head()
```

```python
df['Name'].head(20)
```

```python
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.')
```

```python
df.head()
```

```python
df['Title'].value_counts()
```

```python
df['Title'] = df['Title'].replace({
     'Mr': 'Mr',
    'Mister': 'Mr',
    'Mrs': 'Mrs',
    'Miss': 'Mrs',
    'Ms':'Mrs'
    })
```

```python
df['Title'].value_counts()
```

```python
df['Name'] = df['Name'].str.replace(r'\b(Mr|Mrs|Master|Col|Rev|Dr|Dona|Miss)\.\s*', '', regex=True)
```

```python
df['Name'] = df["Name"].str.replace('"', '', regex=False)
```

```python
df['Name'] = df["Name"].str.replace('(', '', regex=False)
df['Name'] = df["Name"].str.replace(')', '', regex=False)
```

```python
df['Name'] = df['Name'].str.replace(' ', '', regex=False).str.strip()
```

```python
print("Name col is cleaned")
df.tail(7)
```

```python
# surnme and title cols are not needed 

df.drop(['Surname', 'Title'], axis=1, inplace=True)
```

```python
print(df.columns)
print(f"\n The shape of the data is {df.shape}")
```

```python
# starting with the title col

df['Ticket'].value_counts()
```

```python
df['Ticket'] = df['Ticket'].str.replace(r'[^0-9]', '', regex=True)
```
